```python
import matplotlib.pyplot as plt
```

```python
data = [4, 8, 15, 16, 23, 42, 50, 60, 55, 45, 30, 20, 10]
```

```python
def moving_average(data,filter_size):
  new_data = []
  sum=0
  j=0
  for i in range(len(data)):
    if(i<filter_size-1):
      sum +=data[i]
      avg = sum/(i+1)
    else:
      sum +=data[i]
      avg = sum/filter_size
      sum -= data[j]
      j=j+1
    new_data.append(avg)
  return new_data
```

```python
first_order = moving_average(data, 1)  # First-order
second_order = moving_average(data, 2)  # Second order
print(first_order)
print(second_order)
```

```python
x = list(range(len(data)))
plt.figure(figsize=(10, 5))
plt.plot(x, data, label='Original Data', marker='o', linewidth=2)
plt.plot(x, first_order, label='First-order MA (2-point)', marker='x', linestyle='--')
plt.plot(x, second_order, label='Second-order MA (3-point)', marker='s', linestyle=':')

plt.title('Moving Average Filter')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
